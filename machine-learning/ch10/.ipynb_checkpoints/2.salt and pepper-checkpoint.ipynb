```python
%matplotlib inline
import mahotas as mh
from matplotlib import pyplot as plt
import numpy as np
```

```python
lenna = mh.colors.rgb2gray(mh.demos.load('lena'), dtype=np.uint8)
plt.gray()
plt.imshow(lenna)
```

```python
salt = np.random.random(lenna.shape) > .975
pepper = np.random.random(lenna.shape) > .975
```

```python
lenna = mh.stretch(lenna)
lenna = np.maximum(salt * 170, lenna)
lenna = np.minimum(pepper * 30 + lenna*(~pepper), lenna)
```

```python
plt.imshow(lenna)
```

```python
# restor using gaussian filter!
plt.imshow(mh.gaussian_filter(lenna, 1))
```

```python
plt.imshow(mh.gaussian_filter(lenna, 2))
```
