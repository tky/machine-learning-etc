```python
%matplotlib inline
import mahotas as mh
from matplotlib import pyplot as plt
import numpy as np
```

```python
image = mh.colors.rgb2gray(mh.demos.load('lena'), dtype=np.uint8)
```

```python
plt.gray()
plt.imshow(image)
```

```python
filtered = mh.sobel(image, just_filter=True)
```

```python
plt.imshow(filtered)
```

```python
# http://www.scipy-lectures.org/advanced/image_processing/#sharpening
# try sharpening!
from scipy import misc
from scipy import ndimage
```

```python
image = mh.colors.rgb2gray(mh.demos.load('lena'), dtype=float)
blurred_f = ndimage.gaussian_filter(image, 3)
filter_blurred_f = ndimage.gaussian_filter(blurred_f, 1)
```

```python
alpha = 30
sharpened = blurred_f + alpha * (blurred_f - filter_blurred_f)
```

```python
plt.imshow(sharpened)
```

```python
# Ha, difference between  them are verry little. orz.
plt.imshow(mh.sobel(image, just_filter=True))
```
