```python
%matplotlib inline
import mahotas as mh
from matplotlib import pyplot as plt
import numpy as np
```

```python
im = mh.demos.load('lena')
plt.imshow(im)
```

```python
# transpose(転置行列)
r, g, b = im.transpose(2, 0, 1)
```

```python
r12 = mh.gaussian_filter(r, 12.)
g12 = mh.gaussian_filter(g, 12.)
b12 = mh.gaussian_filter(b, 12.)
im12 = mh.as_rgb(r12, g12, b12)
plt.imshow(im12)
```

```python
h, w = r.shape
Y, X = np.mgrid[:h, :w]

Y = Y - h/2.
Y = Y / Y.max()
X = X - w/2.
X = X / X.max()

W = np.exp(-2.*(X**2 + Y**2))
W = W - W.min()
W = W / W.ptp()
W = W[:, :, None]

ringed = mh.stretch(im * W + (1 - W) * im12)
```

```python
plt.imshow(ringed)
```
