```python
%matplotlib inline
import mahotas as mh
from matplotlib import pyplot as plt
import numpy as np
```

```python
orig_image = mh.imread('./SimpleImageDataset/building05.jpg')
```

```python
plt.imshow(orig_image)
```

```python
image = mh.colors.rgb2gray(orig_image, dtype=np.uint8)
```

```python
plt.gray()
plt.imshow(image)
```

```python
thresh = mh.thresholding.otsu(image)
print(thresh)
plt.imshow(image > thresh)
```

```python
otsubin = (image > thresh)
# close small area
otsubin = mh.close(otsubin, np.ones((15, 15)))
plt.imshow(otsubin)
```

```python
# Ridler-Calvard method
print (mh.thresholding.rc(image))
plt.imshow(image > mh.thresholding.rc(image))
```

```python
# Gaussian 
# filter size 8
im8 = mh.gaussian_filter(image, 8)
plt.imshow(im8)
```

```python
plt.imshow(im8 > mh.thresholding.otsu(im8.astype(np.uint8)))
```

```python
# Gaussian 
# filter size 32
im32 = mh.gaussian_filter(image, 32)
plt.imshow(im32)
```

```python
plt.imshow(im32 > mh.thresholding.otsu(im32.astype(np.uint8)))
```
