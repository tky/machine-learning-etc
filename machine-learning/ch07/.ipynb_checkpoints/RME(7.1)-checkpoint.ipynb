```python
%matplotlib inline
from sklearn.datasets import load_boston
from matplotlib import pyplot as plt
import numpy as np
```

```python
boston = load_boston()
```

```python
plt.scatter(boston.data[:,5], boston.target, color='r')
```

```python
x = boston.data[:,5]
x = np.array([[v] for v in x])
y = boston.target
slope,_,_,_ = np.linalg.lstsq(x,y)
```

```python
plt.scatter(boston.data[:,5], boston.target, color='r')
plt.plot([0, boston.data[:, 5].max() + 1],
         [0, slope * (boston.data[:, 5].max() + 1)], '-', lw=4)
plt.show()
```

```python
# 部屋の数に関わりなく最低限必要な価格を設定する。
bx = boston.data[:,5]
bx = np.array([[v,1] for v in bx])
by = boston.target
(bslope,bbias),_,_,_ = np.linalg.lstsq(bx,by)
```

```python
plt.scatter(boston.data[:,5], boston.target, color='r')
plt.plot([0, boston.data[:, 5].max() + 1],
         [0, slope * (boston.data[:, 5].max() + 1)], '-', lw=4)
plt.plot([bx.min(), bx.max()], [bslope * bx.min() + bbias, bslope * bx.max() + bbias], '-', lw=4)
plt.grid()
plt.axis([0, 10, 0, 60])
plt.show()
```

```python
# use all features
ax = boston.data
ax = np.array([np.concatenate((v, [1])) for v in boston.data])
ay = boston.target
_, atotal_error, _, _ = np.linalg.lstsq(ax, ay)

armse = np.sqrt(atotal_error[0] / len(ax))
print('With using all features, Residual: {}'.format(armse))

x = boston.data[:,5]
x = np.array([[v] for v in x])
y = boston.target
_,total_error,_,_ = np.linalg.lstsq(x,y)
rmse = np.sqrt(total_error[0] / len(x))
print('With using only number of rooms , Residual: {}'.format(rmse))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression(fit_intercept=True)
```

```python
x = boston.data
y = boston.target
lr.fit(x, y)
p = np.array(map(lr.predict, x))
p = p.ravel()
e = p - y
total_error = np.sum(e*e)
rmse_train = np.sqrt(total_error/len(p))
print('RMSE on traing: {}'.format(rmse_train))
```

```python
from sklearn.cross_validation import KFold
kf = KFold(len(x), n_folds=10)
err = 0
for train, test in kf:
    lr.fit(x[train], y[train])
    p = map(lr.predict, x[test])
    p = np.array(p).ravel()
    e = p - y[test]
    err += np.dot(e, e)

rmse_10cv = np.sqrt(err / len(x))
print('RMSE on 10-fold CV: {}'.format(rmse_10cv))
```
